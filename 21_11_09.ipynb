```python
def fac(N):
    if N==1 or N==0:
        return 1
    else:
        return N*fac(N-1)
print(fac(5))
```

```python
# 클래스
class Bank:
  def __init__(self):#초기화 설정함수(권한을 줄 때, 자동으로 한번 실행되는 함수)
    print("고객님 가입을 축하드립니다")
    self.money=0 
  def deposit(self,a):
    self.money=a
  def withdraw(self,a)

a=Bank()
```

```python
class Bank2(Bank): #Bank의 기능을 모두 상속 받음
  def withdraw(self,a):
    if self.money-a>0:
      self.money-=a
      print("{}원이 출금되었습니다".format(a))
    else:
      print("잔액이 부족합니다")
b=Bank2()
b.deposit(100000)
b.withdraw(3000)
```
